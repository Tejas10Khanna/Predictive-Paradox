```python
df = df.set_index('datetime').sort_index()
print(df.shape)
print(df.index.min(), "→", df.index.max())
```

```python
df['solar'] = df.apply(
    lambda row: 0 if (row.name.hour < 6 or row.name.hour > 19)
                else row['solar'], axis=1
)
df['solar'] = df['solar'].interpolate(method='time', limit=3).fillna(0)

for col in ['wind', 'india_adani', 'nepal']:
    df[col] = df[col].fillna(0)

print(df[['solar', 'wind', 'india_adani', 'nepal']].isna().sum())
```

```python
roll_mean = df['demand_mw'].rolling(window=24*7, center=True, min_periods=1).mean()
roll_std  = df['demand_mw'].rolling(window=24*7, center=True, min_periods=1).std()
z_score   = (df['demand_mw'] - roll_mean) / roll_std

n_anomalies = (z_score.abs() > 3.5).sum()
print(f"Anomalies detected: {n_anomalies}")
print(f"Max z-score: {z_score.abs().max():.2f}")
```

```python
df.loc[z_score.abs() > 3.5, 'demand_mw'] = np.nan
df['demand_mw'] = df['demand_mw'].interpolate(method='time', limit=3)

print(f"NaNs remaining in demand_mw: {df['demand_mw'].isna().sum()}")
print(f"demand_mw range: {df['demand_mw'].min():.1f} → {df['demand_mw'].max():.1f} MW")
```

```python
melted_eco = economic_df.melt(
    id_vars=['Indicator Name'], 
    value_vars=[str(y) for y in range(2014, 2026)], 
    var_name='year', 
    value_name='Value'
)
annual_eco = melted_eco.pivot(index='year', columns='Indicator Name', values='Value').reset_index()
annual_eco['year'] = annual_eco['year'].astype(int)
df['year'] = df.index.year
df = df.reset_index()
df = pd.merge(df, annual_eco, on='year', how='left')
df = df.set_index('datetime')
df = df.drop(columns=['year'])
print("Merge Successful! Current columns in df:")
print(df.columns)
```

```python
candidate_eco_cols = [col for col in annual_eco.columns if col != 'year']
def select_top_percentile_features(df, target_col, candidate_cols, percentile_threshold=98):
    print(f"Scoring {len(candidate_cols)} features. This might take a few seconds...")
    correlations = df[candidate_cols].corrwith(df[target_col])
    abs_correlations = correlations.abs()
    cutoff_score = np.percentile(abs_correlations.dropna(), percentile_threshold)
    winning_cols = abs_correlations[abs_correlations >= cutoff_score].index.tolist()
    
    print(f"--- Feature Selection Results ---")
    print(f"Score needed to break the {percentile_threshold}th percentile: {cutoff_score:.4f}")
    print(f"Number of columns selected: {len(winning_cols)}\n")
    
    return winning_cols, abs_correlations.sort_values(ascending=False)

top_eco_cols, ranked_scores = select_top_percentile_features(df, 'demand_mw', candidate_eco_cols, percentile_threshold=99)
print("--- Top 10 Strongest Economic Features ---")
print(ranked_scores.head(10))
cols_to_drop = [col for col in candidate_eco_cols if col not in top_eco_cols and col in df.columns]
df = df.drop(columns=cols_to_drop)
print(f"\nTrash removed! Final dataframe shape: {df.shape}")
```

```python
df['hour'] = df['datetime'].dt.hour
df['day_of_week'] = df['datetime'].dt.dayofweek
df['month'] = df['datetime'].dt.month
df['is_weekend'] = df['day_of_week'].isin([4, 5]).astype(int) 
df['demand_lag_1h'] = df['demand_mw'].shift(1)
df['demand_lag_2h'] = df['demand_mw'].shift(2)
df['demand_lag_24h'] = df['demand_mw'].shift(24)
df['demand_lag_168h'] = df['demand_mw'].shift(168) 

df['rolling_mean_24h'] = df['demand_mw'].shift(1).rolling(window=24).mean()
df['rolling_std_24h'] = df['demand_mw'].shift(1).rolling(window=24).std()
df['target_next_hour'] = df['demand_mw'].shift(-1)
df = df.ffill()
df = df.dropna(subset=['demand_lag_168h', 'target_next_hour'])
```

```python
train_df = df[df['datetime'].dt.year <= 2023].copy()
test_df = df[df['datetime'].dt.year == 2024].copy()

target_col = 'target_next_hour'
feature_cols = [col for col in df.columns if col not in [target_col, 'datetime']]
X_train = train_df[feature_cols]
y_train = train_df[target_col]

X_test = test_df[feature_cols]
y_test = test_df[target_col]

print(f"Training Features (X_train) shape: {X_train.shape}")
print(f"Testing Features (X_test) shape: {X_test.shape}")
```

```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
plt.style.use('ggplot')
xgb_model = xgb.XGBRegressor(
    n_estimators=500,        
    learning_rate=0.05,      
    max_depth=6,            
    random_state=42,         
    n_jobs=-1                
)
xgb_model.fit(X_train, y_train)
predictions = xgb_model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
mape = np.mean(np.abs((y_test - predictions) / y_test)) * 100

print("\n--- Model Performance (2024 Test Set) ---")
print(f"Mean Absolute Error (MAE): {mae:.1f} MW")
print(f"Root Mean Squared Error (RMSE): {rmse:.1f} MW")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
hours_to_plot = 336 

plt.figure(figsize=(16, 6))
plt.plot(test_df['datetime'].iloc[:hours_to_plot], y_test.iloc[:hours_to_plot], 
         label='Actual Demand', color='blue', linewidth=2)
plt.plot(test_df['datetime'].iloc[:hours_to_plot], predictions[:hours_to_plot], 
         label='XGBoost Predicted Demand', color='orange', linestyle='--', linewidth=2)

plt.title('Time-Series Forecast: Actual vs. Predicted Power Demand (First 2 Weeks of 2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Power Demand (MW)', fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
